## Saving dataframes for post-analysis

```python
# Performing a merge on the 'wise.id' column
merged_df = pd.merge(df_wise, df_platform_wise[['wise.id', 'id']], on='wise.id', how='left')

# Renaming the 'id' column to 'platform_id' and keeping the original columns of df_wise
df_wise['platform_id'] = merged_df['id']

# Performing a merge on the 'Transaction ID' column
merged_df = pd.merge(df_paypal, df_platform_paypal[['Transaction ID', 'id']], on='Transaction ID', how='left')

# Renaming the 'id' column to 'platform_id' and keeping the original columns of df_paypal
df_paypal['platform_id'] = merged_df['id']

# Performing a merge on the 'stripe.id' column
merged_df = pd.merge(df_stripe, df_platform_stripe[['stripe.id', 'id']], on='stripe.id', how='left')

# Renaming the 'id' column to 'platform_id' and keeping the original columns of df_stripe
df_stripe['platform_id'] = merged_df['id']

# Performing a merge on the 'index' column
merged_df = pd.merge(df_banks, df_platform_bank[['index', 'id']], on='index', how='left')

# Renaming the 'id' column to 'platform_id' and keeping the original columns of df_banks
df_banks['platform_id'] = merged_df['id']

report_data = {
    'wise': df_wise,
    'paypal': df_paypal,
    'stripe': df_stripe,
    'bank': df_banks,
    'platform_wise': df_platform_wise,
    'platform_paypal': df_platform_paypal,
    'platform_stripe': df_platform_stripe,
    'platform_bank': df_platform_bank,
    'left_bank': df_left_bank,
    'left_wise': df_left_wise,
    'left_stripe': df_left_stripe,
    'left_paypal': df_left_paypal,
    'platform': df_platform
}

import pickle

# save report_data to a pickle file in data folder
with open('reports/opensource_report_data.pickle', 'wb') as f:
    pickle.dump(report_data, f)
```
